# Funciones generales

````python
def get_image(level: int):
  if not level or level > 5:
    raise Exception(f"Level {level} not found.")

  if level == 1:
    image1 = np.ones((5, 5))
    image1[:, 1:3] = 0

    image2 = np.zeros((5, 5))

  elif level == 2:
    image1 = np.ones((5, 5))
    image1[:2, 3:] = 0
    image1[4, :] = 0
    image1[:1, :1] = 0

    image2 = np.ones((5, 5))
    image2[3, :3] = 0

  elif level == 3:
    image1 = np.ones((5, 5))
    image1[:2, :2] = 0
    image1[4:, 3:] = 0

    image2 = np.ones((5, 5))
    image2[3, 3] = 0
    image2[4:, 4:] = 0

  elif level == 4:
    image1 = np.ones((7, 7))
    image1[:3, :] = 0
    image1[:, 6:] = 0
    image1[6, 6] = 0

    image2 = np.identity(7)

  elif level == 5:
    image1 = np.ones((7, 7))
    image1[:5, :6] = 0
    image1[:, :3] = 0

    image2 = np.ones((7, 7))
    image2[:3, :3] = 0
    image2[5, 5:7] = 0

  return [(image1 * 255).astype(np.uint8), (image2 * 255).astype(np.uint8)]

def show_image(image: np.array):
  plt.imshow(image, cmap="binary");
````

image.png

````python
image1, image2 = get_image(3)

image = cv2.bitwise_not(image2)
image = cv2.bitwise_and(image, image1)

show_image(image)
````

````python
image1, image2 = get_image(4)

image = cv2.bitwise_not(image1)
image = cv2.bitwise_xor(image2, image)
image = cv2.rotate(image, 1)
image = cv2.bitwise_or(image, image2)

show_image(image)
````
